### What is Numpy? and How to install it ?

- NumPy --> Numerical Python

- Numpy is the core library for scientific computing in Python. 

- It provides a high-performance multidimensional array object, and tools for working with these arrays.

- Array : A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers. The number of dimensions is the rank of the array; the shape of an array is a tuple of integers giving the size of the array along each dimension.
- To install it : pip install numpy 
- Then need to import it in the jupiter notebook.

### Creating Array

```python
# Creating numpy array using Python List 
# 1D arrary

myList = [1,2,3,4,5]

print("myList: ", myList)
print("Type: ", type(myList))
print("No of element: ", len(myList))
print("-"*25)

array1 = np.array(myList)

print("array1: ", array1)
print("Type: ", type(array1))
print("No of elements: ", array1.size)
print("-"*25)
```

```python
# Creating numpy array using Python List 
# 2D arrary

myList = [[1,2,3],
          [4,5,6],
          [7,8,9]]

print("myList\n", myList)
print("-"*25)

array1_2D = np.array(myList)
print("aaray1_2D\n", array1_2D)
print("No of elements: ",array1_2D.size) 
print("Type: ", type(array1_2D))
# Check te dimension
print("Dimension of Arrary1_2D: ", array1_2D.ndim)

# Shape of array
print("Shape of the arrary1_2D: ", array1_2D.shape)
```

```python
# Check the datatype

print("Data Type: ",array1_2D.dtype)
```

```python
# We can also create a numpy array with float

myList = [[1,2,3],
          [4,5,6],
          [7,8,9]]      # 2D list

print("myList\n", myList)
print("-"*35)

array1_2D = np.array(myList, dtype=np.float32)
print("aaray1_2D\n", array1_2D)
print("No of elements: ",array1_2D.size)
print("Type: ", type(array1_2D)) 

# Check te dimension
print("Dimension of Arrary1_2D: ", array1_2D.ndim)

# Shape of array
print("Shape of the arrary1_2D: ", array1_2D.shape)
```

```python
# Creating from tuple


myTuple = (1,2,3,4,5)

print("myTuple: ", myTuple)
print("Type: ", type(myTuple))
print("No of element: ", len(myTuple))
print("-"*25)

array1 = np.array(myTuple)

print("array1: ", array1)
print("Type: ", type(array1))
print("No of elements: ", array1.size)
print("-"*25)
```

```python
# Creating from tuple - 2D


myTuple2D = ((1,2,3),(4,5,6),(7,8,9))

print("myTuple: ", myTuple2D)
print("Type: ", type(myTuple2D))
print("No of element: ", len(myTuple2D))
print("-"*25)

array1_2D = np.array(myTuple2D)

print("array1: ", array1_2D)
print("Type: ", type(array1_2D))
print("No of elements: ", array1_2D.size)
print("Dimension: ", array1_2D.ndim)
print("-"*25)
```

```python
# Creating the numpy with range
# np.arange(start, stop, step)

array2 = np.arange(0, 12)
print("array2: ", array2)
print("Type: ", type(array2))
print("Dim: ", array2.ndim)
print("Number of element: ", array2.size)
print("shape: ", array2.shape)
```

```python
# You can reshape in 2D

array2 = np.arange(0, 12).reshape(3,4)
print("array2: ", array2)
print("Type: ", type(array2))
print("Dim: ", array2.ndim)
print("Number of element: ", array2.size)
print("shape: ", array2.shape)
```

```python
# Linspace

array3 = np.linspace(0, 10,5, dtype=np.int32)
print("array2: ", array3)
print("Type: ", type(array3))
print("Dim: ", array3.ndim)
print("Number of element: ", array3.size)
print("shape: ", array3.shape)
```

```python
# Create an array of random values of given shape

arrary_3 = np.random.rand(3,3) # Uniform distribution
print("array_3: ", arrary_3)
print("Size: ", arrary_3.size)
print("shape: ", arrary_3.shape)
print("Dim: ", arrary_3.ndim)
```

```python
# Create an array of random values of given shape

arrary_3 = np.random.randn(3,3) # Normal distribution
print("array_3: ", arrary_3)
print("Size: ", arrary_3.size)
print("shape: ", arrary_3.shape)
print("Dim: ", arrary_3.ndim)
```

```python
# Zeros

array_zeros = np.zeros((3,3), dtype=np.int32)
print("array_3: ", array_zeros)
print("Size: ", array_zeros.size)
print("shape: ", array_zeros.shape)
print("Dim: ", array_zeros.ndim)
```

```python
# Ones

array_zeros = np.ones((3,3), dtype=np.int32)
print("array_3: ", array_zeros)
print("Size: ", array_zeros.size)
print("shape: ", array_zeros.shape)
print("Dim: ", array_zeros.ndim)
```

```python
# Eye

array_zeros = np.eye(3, dtype=np.int32)
print("array_3: ", array_zeros)
print("Size: ", array_zeros.size)
print("shape: ", array_zeros.shape)
print("Dim: ", array_zeros.ndim)
```

### Basic Operations

```python
a = np.arange(1,13).reshape(3,4)

print("a: \n", a)

b = np.arange(20,32).reshape(3,4)

print("b: \n", b)
```

```python
# Addition

c = a + b
print("c:\n", c)
```

```python
# Subtractions

c = a - b
print("c:\n", c)
```

```python
# Multiplication by scalar

c = a*2
print("c:\n", c)
```

```python
# Multiplication of two 2D numpy array
# # Element-wise multiplication
c = a*b
print("c:\n", c)
```

```python
# Element-wise division

c = b/a
print("c:\n", c)
```

```python
# Dot product

c = np.dot(a,b)

print("Dot product of a and b:\n", c)
```
